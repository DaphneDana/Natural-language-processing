```python
#Bag of words
class Category:
    BOOKS = "BOOKS"
    CLOTHING = "CLOTHING"
train_x = ["I love the book", "this is a great book", "the fit is great", " i love the shoes"]
train_y = [Category.BOOKS, Category.BOOKS, Category.CLOTHING, Category.CLOTHING]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary = True)  # you can use bi grams too if you have positive and negative statements
train_x_vectors = vectorizer.fit_transform(train_x)  # fits the words in the dictornary to pick the unique words and then tokenize them
print (train_x_vectors.toarray())
print (vectorizer.get_feature_names_out())
```

```python
# use the linear svm model becaus eit is good with text classification
from sklearn import svm
clf_svm = svm.SVC(kernel = 'linear')
clf_svm.fit(train_x_vectors, train_y)
```

```python
test_x = vectorizer.transform(['i like the book'])
clf_svm.predict(test_x)
```

```python
test_x = vectorizer.transform(['shoes are alright'])
clf_svm.predict(test_x)
```

```python
test_x = vectorizer.transform(['bags are ugly'])
clf_svm.predict(test_x)
```

NICE TO HAVES IN THE nlp CODE

REGEXES: Pattern matching of strings in python .             Password checkers, phone numbers , emails and more


```python
import re

regexp = re.compile(r"^ab[^\s]*cd$")

phrases = ["abcd", "xxx", "abxxxcd", "ab cd"]

matches = []
for phrase in phrases:
    if regexp.match(phrase):   # or re.match(regexp, phrase) also works
        matches.append(phrase)

print(matches)
```

```python
import re

regexp = re.compile(r"read|story|book")

phrases = ["I like that book", "I like to read", "this hat is nice"]

matches = []
for phrase in phrases:
    if re.search(regexp, phrase):   # or re.match(regexp, phrase) also works
        matches.append(phrase)

print(matches)
```

STEMMING/ LEMMETAIZATION

techniques to normalize text

```python
#nltk : Natural language toolkit
import nltk
nltk.download ('wordnet')
nltk.download ('stopwords')
nltk.download ('punkt')
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize  # can take a sentence and break it into its individial words
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

phrase = "reading the books"
words = word_tokenize(phrase)

stemmed_words = []
for word in words:
    stemmed_words.append(stemmer.stem(word))

" ".join(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer ()

phrase = " reading the books"
words = word_tokenize(phrase)

lemmatized_words = []
for word in words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

" ".join(lemmatized_words)
```

STOPWORDS REMOVAL
These are set of most common words in english eg This, that, he , it

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = stopwords.words("english")
print (stop_words)

phrase = "the cat is on the table"
words = word_tokenize(phrase)

stripped_phrase = []
for word in words:
    if word not in stop_words:
        stripped_phrase.append(word)

" ".join(stripped_phrase)
```
